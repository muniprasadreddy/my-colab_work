# preprocessing

```python
import os
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.transform import resize
```

```python


target = []
images = []
flat_data = []

Datadir = '/content/images'
Categories = ['different pretty flowers', 'fruits and vegetables', 'ice cream']

for cate in Categories:
    class_num = Categories.index(cate) ### Label encoding the values
    print(class_num)
    path = os.path.join(Datadir, cate)    # create path to use all the images
    print(path)
```

```python
from skimage.io import imread
from skimage.transform import resize
import numpy as np

flat_data = []
images = []
target = []

for img in os.listdir(path):
    img_arry = imread(os.path.join(path, img))
    #print( img_arry.shape)
    #plt.imshow(img_arry)
                  ###### resize
### flatten
    img_resized = resize(img_arry, (200, 200, 3)) ## Normalized thr value from 0 to1
    flat_data.append(img_resized.flatten())
    images.append(img_resized)
    target.append(class_num)


flat_data = np.vstack(flat_data)
target = np.array(target)
images = np.array(images)
```
